RENOMEAR COLUNAS PARA FACILITAR

```python
#rename columns 

books = books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]

books.rename(columns = {'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'years', 'Publisher':'publisher'}, inplace=True)
users.rename(columns = {'User-ID':'user_id', 'Location':'location', 'Age':'age'}, inplace=True)
rating.rename(columns = {'User-ID':'user_id', 'Book-Rating':'rating'}, inplace=True)
```

```python
#quantidade de Ranting dos Usuários, essencial para um analista de dados e cientista nesse caso de recomendação 

rating['user_id'].value_counts()
```

```python
#esse código foi feito para filtrar as pessoas que avaliaram mais de 200 livros

x = rating['user_id'].value_counts() > 200

x
```

```python
#quantidade de Usuários

y = x[x].index

print(y.shape)
```

```python
rating[rating['rating'] > 200 ]
```

```python
#(y) é onde todos os que avaliaram estão 

y
```

```python
#trazendo ratings somentes dos usuarios que avaliaram mais de 200 livros do (y)

rating = rating[rating['user_id'].isin(y)]
```

```python
# '.merge' é a mesclação de duas variavéis de rating and books

rating_with_books = rating.merge(books, on='ISBN')
rating_with_books
```

```python
number_rating = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating = number_rating.rename(columns = {'rating':'Number_of_ratings'})
```

```python
number_rating
```

```python
#juntando as tabelas de livros com ratings com a tabela de quantidade cada livro

final_rating = rating_with_books.merge(number_rating, on='title')
final_rating
```

```python
final_rating = final_rating[final_rating['Number_of_ratings'] >= 50]
final_rating.shape
```
