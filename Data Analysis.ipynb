```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = pd.read_csv('SuperStoreOrders.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#isna is to know is valuer is NAN, mean() function can be used to calculate mean/average of a given list of numbers. 

df.isna().mean()
```

```python
#use Sort_values to sorted columns 

df = df.sort_values(by=['profit'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['order_date'] = df['order_date'].astype('datetime64[ns]')
df['ship_date'] = df['ship_date'].astype('datetime64[ns]')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
```

```python
df.dtypes
```

```python
#Check for Null Values
df.isnull().sum()
```

```python
df['sales'] = df['sales'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.describe()
```
