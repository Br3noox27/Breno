```python
import requests

from bs4 import BeautifulSoup

lista_de_produtos = []

numero_de_paginas = 1

headers = {'user-agent': 'mozilla/5.0'}

url = 'https://sp.olx.com.br/?q=iphone'

oq_vc_qr = input('o que vc deseja ? ')

while numero_de_paginas > 0:

  url_ = requests.get(f'https://sp.olx.com.br/?o={numero_de_paginas}&q={oq_vc_qr}', headers=headers)

  site = BeautifulSoup(url_.text, 'html.parser')

  produtos = site.findAll('li', attrs={'class':'sc-1fcmfeb-2 fvbmlV'})

  for produto in produtos:
    try:

      nome = produto.find('h2', attrs={'class':'kgl1mq-0 eFXRHn sc-ifAKCX gzAEjc'})
      preco = produto.find('span', attrs={'class':'m7nrfa-0 eJCbzj sc-ifAKCX ANnoQ'})
      lugar = produto.find('span', attrs={'class':'sc-1c3ysll-1 cLQXSQ sc-ifAKCX fCbscF'})
      data = produto.find('span', attrs={'class':'sc-11h4wdr-0 javKJU sc-ifAKCX fCbscF'})


      try:
        link = produto.find('a', attrs={'sc-12rk7z2-1 huFwya sc-htoDjs fpYhGm'})['href']

      except: 
        link(' ')

      print(nome.text)
      print(preco.text)
      print(lugar.text)
      print(data.text)
      print(link)

      dicionario = {
          'TITULO DO PRODUTO: ', nome.text,
          'PREÇO DO PRODUTO: ', preco.text,
          'LUGAR DO VENDEDOR: ', lugar.text,
          'DATA DA POSTAGEM: ', data.text,
          'LINK DO PRODUTO: ', link
      }
      lista_de_produtos.append(dicionario)

      print(' ')

    except: ' '
numero_de_paginas += 1
```
